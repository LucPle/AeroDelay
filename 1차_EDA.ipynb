```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rc('font', family='NanumBarunGothic') 
```

```python
path = './x_train_1st.csv'
df = pd.read_csv(path)
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
test = pd.DataFrame()
```

```python
test['날짜'] = df.apply(lambda x: f"{x['년']}{str(x['월']).zfill(2)}{str(x['일']).zfill(2)}", axis=1)
test['날짜'] = test['날짜'].astype('int')
```

```python
test['항공사'] = df['항공사'].copy()
test['편명'] = df['편명'].copy()
test['출발지'] = df['출발지'].copy()
```

```python
test['예상'] = df.apply(lambda x: f"{str(x['예상_1'])}{str(x['예상_2']).zfill(2)}", axis=1)
test['예상'] = test['예상'].astype('int')
test['도착'] = df.apply(lambda x: f"{str(x['도착_1'])}{str(x['도착_2']).zfill(2)}", axis=1)
test['도착'] = test['도착'].astype('int')
```

```python
test['현황'] = df['현황'].copy()
```
